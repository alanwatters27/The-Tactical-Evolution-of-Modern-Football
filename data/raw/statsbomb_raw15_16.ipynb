```python
def extract_team_formations(lineups_raw, match_id):
    rows = []
    for team in lineups_raw:
        team_name = team.get("team_name") or team.get("team", {}).get("name")
        tactics = team.get("tactics") or {}
        formation = tactics.get("formation")  # often an int like 433, 4231, etc.

        rows.append({
            "match_id": match_id,
            "team_name": team_name,
            "formation_raw": formation})
    return pd.DataFrame(rows)

team_formations = extract_team_formations(lineups_raw, match_id)
team_formations
```

Formations were extracted from the Starting XI event in the StatsBomb data and put into a standard hyphenated structure.

```python
import json
import time
import pandas as pd
from urllib.request import urlopen

BASE_URL = "https://raw.githubusercontent.com/statsbomb/open-data/master/data"

def read_json(url):
    with urlopen(url) as r:
        return json.loads(r.read().decode("utf-8"))

def clean_formation(f):
    if pd.isna(f):
        return "Unknown"
    f = str(int(f))          
    return "-".join(f)       
```

```python
pl_match_ids = pl_matches["match_id"].astype(int).tolist()
bund_match_ids = bund_matches["match_id"].astype(int).tolist()

match_ids = pl_match_ids + bund_match_ids
len(match_ids)
```

```python
def extract_formations_from_events(match_id: int) -> pd.DataFrame:
    events_raw = read_json(f"{BASE_URL}/events/{match_id}.json")
    ev = pd.json_normalize(events_raw)
    ev["match_id"] = match_id

    sx = ev[ev["type.name"] == "Starting XI"].copy()

    formations = (
        sx[["match_id", "team.name", "tactics.formation"]].drop_duplicates().rename(columns={"team.name": "team_name", "tactics.formation": "formation_raw"}))

    formations["formation"] = formations["formation_raw"].apply(clean_formation)
    formations = formations.drop(columns=["formation_raw"])
    return formations

formations_all = []
errors_form = []

for i, mid in enumerate(match_ids, start=1):
    try:
        df = extract_formations_from_events(mid)
        formations_all.append(df)

        if i % 25 == 0:
            print(f"Formations: processed {i}/{len(match_ids)} matches")

        time.sleep(0.05)  

    except Exception as e:
        errors_form.append((mid, str(e)))

team_formations_all = pd.concat(formations_all, ignore_index=True)
team_formations_all.head()
```

```python
def extract_shots_from_events(match_id: int) -> pd.DataFrame:
    events_raw = read_json(f"{BASE_URL}/events/{match_id}.json")
    ev = pd.json_normalize(events_raw)
    ev["match_id"] = match_id

    shots = ev[ev["type.name"] == "Shot"].copy()

    
    shots["x"] = shots["location"].apply(lambda v: v[0] if isinstance(v, list) and len(v) > 0 else None)
    shots["y"] = shots["location"].apply(lambda v: v[1] if isinstance(v, list) and len(v) > 1 else None)

    
    shots["end_x"] = shots["shot.end_location"].apply(lambda v: v[0] if isinstance(v, list) and len(v) > 0 else None)
    shots["end_y"] = shots["shot.end_location"].apply(lambda v: v[1] if isinstance(v, list) and len(v) > 1 else None)
    shots["end_z"] = shots["shot.end_location"].apply(lambda v: v[2] if isinstance(v, list) and len(v) > 2 else None)

    keep = shots[[
        "id", "match_id", "minute", "second",
        "team.name", "player.name",
        "type.name",
        "shot.statsbomb_xg",
        "shot.outcome.name",
        "shot.type.name",
        "shot.body_part.name",
        "shot.technique.name",
        "play_pattern.name",
        "x", "y", "end_x", "end_y", "end_z"]].copy()

    keep = keep.rename(columns={
        "id": "event_id",
        "team.name": "team_name",
        "player.name": "player_name",
        "type.name": "event_type",
        "shot.statsbomb_xg": "xg",
        "shot.outcome.name": "shot_outcome",
        "shot.type.name": "shot_type",
        "shot.body_part.name": "body_part",
        "shot.technique.name": "technique",
        "play_pattern.name": "play_pattern"})

    
    keep["xg"] = keep["xg"].fillna(0)
    keep["shot_type"] = keep["shot_type"].fillna("Unknown")
    keep["shot_outcome"] = keep["shot_outcome"].fillna("Unknown")
    keep["is_goal"] = (keep["shot_outcome"] == "Goal").astype(int)

    return keep

shots_all = []
errors_shots = []

for i, mid in enumerate(match_ids, start=1):
    try:
        df = extract_shots_from_events(mid)
        shots_all.append(df)

        if i % 25 == 0:
            print(f"Shots: processed {i}/{len(match_ids)} matches")

        time.sleep(0.05)

    except Exception as e:
        errors_shots.append((mid, str(e)))

shots_all_df = pd.concat(shots_all, ignore_index=True)
shots_all_df.head()
```
