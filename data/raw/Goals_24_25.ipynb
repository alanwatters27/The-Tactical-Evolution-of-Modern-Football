```python
import pandas as pd

pl_url = "https://www.football-data.co.uk/mmz4281/2425/E0.csv"
bund_url = "https://www.football-data.co.uk/mmz4281/2425/D1.csv"

pl_matches = pd.read_csv(pl_url)
bund_matches = pd.read_csv(bund_url)

pl_matches.head()
```

```python
def clean_goals(df, league_name):
    df = df.copy()
    df["total_goals"] = df["FTHG"] + df["FTAG"]
    df["league"] = league_name
    return df[["league","Date","HomeTeam","AwayTeam","FTHG","FTAG","total_goals"]]

pl_goals = clean_goals(pl_matches, "Premier League")
bund_goals = clean_goals(bund_matches, "Bundesliga")

goals_df = pd.concat([pl_goals, bund_goals], ignore_index=True)
goals_df.head()
```

```python
goals_df.to_csv(
    "../cleaned/goals_2425_pl_bundesliga.csv",
    index=False)
```
